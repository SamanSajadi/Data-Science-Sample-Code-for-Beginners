```python
import pandas as pd

# Create a dictionary with data
data = {
    'Name': ['John', 'Jane', 'Mike', 'Lisa'],
    'Age': [25, 30, 35, 28],
    'City': ['New York', 'Paris', 'London', 'Tokyo']
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('dataset.csv', index=False)
```

```python

# Read the CSV file into a DataFrame
df2 = pd.read_csv('dataset.csv')

# Display the DataFrame
print(df2)
```

```python
# Select a single column
n = df['Name']
#print (n)
# Select multiple columns
subset = df[['Name', 'Age']]
print(subset)
```

```python
# Filter rows based on a condition
filtered_df = df[df['Age'] > 30]
print(filtered_df)
```

```python
# Add a new column
df['Country'] = ['USA', 'France', 'UK', 'Japan']
print( df)
```

```python
# Calculate the average age
average_age = df['Age'].mean()
print(average_age)
```

```python
# Sort the DataFrame by a column in ascending order
sorted_df = df.sort_values('Age')
#print(sorted_df)
# Sort the DataFrame by multiple columns in descending order
sorted_df = df.sort_values(['Age', 'Name'], ascending=[False, True])
```

```python
# Group the DataFrame by a column and calculate multiple aggregations
grouped_df = df.groupby('City').agg({'Age': 'mean', 'Name': 'count'})
print(grouped_df)
# Reset the index of the grouped DataFrame
grouped_df = grouped_df.reset_index()
```
