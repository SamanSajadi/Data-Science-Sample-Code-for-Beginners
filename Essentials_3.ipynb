# **Line Plot**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some medical data
time = np.arange(0, 10, 0.1)
heart_rate = 60 + np.sin(time)

# Create a line plot for heart rate over time
plt.plot( heart_rate)

# Add labels and title
plt.xlabel('Time (s)')
plt.ylabel('Heart Rate (bpm)')
plt.title('Heart Rate Variation')

# Display the plot
plt.show()
```

# **Scatter Plot**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate medical data
height = np.random.normal(170, 10, 100)
weight = np.random.normal(70, 5, 100)
age = np.random.randint(18, 65, 100)

# Create a scatter plot for height and weight
plt.scatter(height, weight)

# Add labels and title
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight')

# Display the plot
plt.show()
```

# **Histogram**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate medical data
cholesterol = np.random.normal(200, 30, 500)

# Create a histogram for cholesterol levels
plt.hist(cholesterol, bins=30, edgecolor='black')

# Add labels and title
plt.xlabel('Cholesterol Level (mg/dL)')
plt.ylabel('Frequency')
plt.title('Cholesterol Distribution')

# Display the plot
plt.show()
```

# **Box Plot**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate medical data
bp_systolic = np.random.normal(120, 10, 100)
bp_diastolic = np.random.normal(80, 5, 100)
bp_pulse_pressure = bp_systolic - bp_diastolic

# Create a box plot for pulse pressure
plt.boxplot(bp_pulse_pressure)

# Add labels and title
plt.ylabel('Pulse Pressure (mmHg)')
plt.title('Distribution of Pulse Pressure')

# Display the plot
plt.show()
```

# **Heatmap**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate medical data
correlation_matrix = np.random.rand(5, 5)

# Create a heatmap for correlation matrix
plt.imshow(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)

# Add colorbar
plt.colorbar()

# Add labels and title
plt.xlabel('Variable')
plt.ylabel('Variable')
plt.title('Correlation Matrix')

# Display the plot
plt.show()
```

# **Seaborn **

# **Box Plot**

```python
import seaborn as sns
import numpy as np

# Generate medical data
bp_systolic = np.random.normal(120, 10, 100)
bp_diastolic = np.random.normal(80, 5, 100)
bp_pulse_pressure = bp_systolic - bp_diastolic
gender = np.random.choice(['Male', 'Female'], 100)

# Create a box plot for pulse pressure by gender
sns.boxplot(x=gender, y=bp_pulse_pressure)

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Pulse Pressure (mmHg)')
plt.title('Pulse Pressure by Gender')

# Display the plot
plt.show()
```

# **Scatter Plot**

```python
import seaborn as sns
import numpy as np

# Generate medical data
height = np.random.normal(170, 10, 100)
weight = np.random.normal(70, 5, 100)
age = np.random.randint(18, 65, 100)
gender = np.random.choice(['Male', 'Female'], 100)

# Create a scatter plot for height and weight, colored by gender
sns.scatterplot(x=height, y=weight, hue=gender)

# Add labels and title
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight')

# Display the plot
plt.show()
```

# **Violin Plot**

```python
import seaborn as sns
import numpy as np

# Generate medical data
age_group = np.random.choice(['18-30', '31-45', '46-60', '61+'], 100)
cholesterol = np.random.normal(200, 30, 100)

# Create a violin plot for cholesterol levels by age group
sns.violinplot(x=age_group, y=cholesterol)

# Add labels and title
plt.xlabel('Age Group')
plt.ylabel('Cholesterol Level (mg/dL)')
plt.title('Cholesterol Levels by Age Group')

# Display the plot
plt.show()
```

# **Heatmap**

```python
import seaborn as sns
import numpy as np

# Generate medical data
correlation_matrix = np.random.rand(5, 5)

# Create a heatmap for correlation matrix
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True)

# Add labels and title
plt.xlabel('Variable')
plt.ylabel('Variable')
plt.title('Correlation Matrix')

# Display the plot
plt.show()
```

# **Count Plot**

```python
import seaborn as sns
import numpy as np

# Generate medical data
diagnosis = np.random.choice(['Healthy', 'Disease'], 100)

# Create a count plot for diagnosis
sns.countplot(x=diagnosis)

# Add labels and title
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.title('Diagnosis Distribution')

# Display the plot
plt.show()
```
