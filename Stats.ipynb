# Importing the necessary libraries

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

# **Generating sample data:**
Let's assume we have a medical dataset with two groups (Group A and Group B)

```python
np.random.seed(0)
group_a = np.random.normal(loc=10, scale=2, size=100)
group_b = np.random.normal(loc=12, scale=2, size=100)


# Descriptive statistics

print("Group A mean:", np.mean(group_a))
print("Group B mean:", np.mean(group_b))
print("Group A standard deviation:", np.std(group_a))
print("Group B standard deviation:", np.std(group_b))

# Visualization
plt.hist(group_a, alpha=0.5, label='Group A', bins=15)
plt.hist(group_b, alpha=0.5, label='Group B', bins=15)
plt.legend(loc='best')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of Group A and Group B')
plt.show()
```

We can use **t-tests** to compare the means of the two groups

```python
# Independent two-sample t-test
t_stat, p_value = stats.ttest_ind(group_a, group_b, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)


plt.boxplot([group_a, group_b], labels=['Group A', 'Group B'])
plt.ylabel('Value')
plt.title('Boxplot of Group A and Group B')
plt.show()
```

**# Mann-Whitney U test **(for non-parametric comparisons) **or chi-square tests** (for categorical data).

```python
import scipy.stats as stats

# Generating sample data
group_a = [23, 25, 18, 20, 21, 19, 17, 24, 22, 16]
group_b = [27, 29, 28, 26, 31, 30, 32, 34, 33, 35]

# Perform Mann-Whitney U test
u_stat, p_value = stats.mannwhitneyu(group_a, group_b)

print("U statistic:", u_stat)
print("P-value:", p_value)
```

```python
import scipy.stats as stats

# Generating sample data
observed = [[10, 15, 25],
            [20, 10, 15]]

# Perform chi-square test
chi2_stat, p_value = stats.chisquare(observed)

print("Chi-square statistic:", chi2_stat)
print("P-value:", p_value)

```

# **ANOVA is used to compare the means of three or more groups.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(0)
group_a = np.random.normal(loc=10, scale=2, size=100)
group_b = np.random.normal(loc=12, scale=2, size=100)
group_c = np.random.normal(loc=15, scale=2, size=100)
```

```python
# Descriptive statistics
print("Group A mean:", np.mean(group_a))
print("Group B mean:", np.mean(group_b))
print("Group C mean:", np.mean(group_c))

# Visualization
plt.boxplot([group_a, group_b, group_c], labels=['Group A', 'Group B', 'Group C'])
plt.ylabel('Value')
plt.title('Boxplot of Group A, Group B, and Group C')
plt.show()
```

```python
f_stat, p_value = stats.f_oneway(group_a, group_b, group_c)

print("F-statistic:", f_stat)
print("P-value:", p_value)
```

**If the ANOVA test yields a significant result (i.e., p-value < 0.05), you can perform post hoc tests to determine which specific group differences are significant. **

```python
means = [np.mean(group_a), np.mean(group_b), np.mean(group_c)]
errors = [np.std(group_a), np.std(group_b), np.std(group_c)]

plt.bar(['Group A', 'Group B', 'Group C'], means, yerr=errors)
plt.ylabel('Mean')
plt.title('Mean Comparison of Group A, Group B, and Group C')
plt.show()
```
